```python
import matplotlib.pyplot as plt
import numpy as np

# Data for "Bertha"
beats_bertha = np.arange(1, 664)  # Number of beats (1 to 663)
tempo_bertha = np.random.normal(118.7, 5.2, len(beats_bertha))  # Random data centered around the average tempo with some standard deviation

# Data for "CCS/IKYR"
beats_ccs_ikyr = np.arange(1, 947)  # Number of beats (1 to 946)
tempo_ccs_ikyr = np.random.normal(90.0, 3.7, len(beats_ccs_ikyr))  # Random data centered around the average tempo with some standard deviation

# Create the scatter plot
plt.figure(figsize=(12, 6))

# Scatter plot for "Bertha"
plt.scatter(beats_bertha, tempo_bertha, color='blue', alpha=0.5, label='Bertha: Faster and More Variable')

# Scatter plot for "CCS/IKYR"
plt.scatter(beats_ccs_ikyr, tempo_ccs_ikyr, color='green', alpha=0.5, label='CCS/IKYR: Slower and More Consistent')

# Adding titles and labels
plt.title('Tempo Variation Across Beats for "Bertha" and "CCS/IKYR"')
plt.xlabel('Beat Number')
plt.ylabel('Tempo (BPM)')
plt.legend()

# Show the plot
plt.show()
```
